Exploratory Data Analysis (EDA): 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset from a text file
data = np.loadtxt('ECG200_TRAIN.txt')

# Check the first few rows of the dataset
print("First row of the dataset:")
print(data[:1])

# Convert the data into a DataFrame
df = pd.DataFrame(data)

# Check the distribution of classes
class_distribution = df[0].value_counts()
print("\nDistribution of classes:")
print(class_distribution)

# Visualize class distribution
plt.figure(figsize=(8, 6))
sns.countplot(x=df[0])
plt.title('Class Distribution')
plt.xlabel('Class')
plt.ylabel('Count')
plt.show()

# Check for class imbalance
if len(class_distribution) > 1:
    imbalance_ratio = class_distribution.min() / class_distribution.max()
    print("\nClass imbalance ratio:", imbalance_ratio)

# Visualize sample ECG recordings
plt.figure(figsize=(10, 6))
for i in range(5):  # Plotting 5 sample ECG recordings
    plt.subplot(2, 3, i+1)
    plt.plot(df.iloc[i, 1:])
    plt.title('ECG Sample {}'.format(i+1))
    plt.xlabel('Time')
    plt.ylabel('Signal')
plt.tight_layout()
plt.show()
```

Class imbalance ratio: 0.4492753623188406
so that we need to balance it

using upsampling: random oversampling:
This process increases the number of instances in the minority class by randomly duplicating samples, effectively balancing the class distribution in the dataset.

```python
from sklearn.utils import resample

# Upsample minority class
minority_upsampled = resample(df[df[0] == -1], replace=True, n_samples=len(df[df[0] == 1]))

# Combine majority class with upsampled minority class
df_balanced = pd.concat([df[df[0] == 1], minority_upsampled])

print("\nClass imbalance detected. Resampling to balance classes...")
print("Original class distribution:")
print(df[0].value_counts())
print("\nBalanced class distribution:")
print(df_balanced[0].value_counts())

# Separate features and labels
df = df_balanced.drop(0, axis=1)
y = df_balanced[0]

```

feature engineering

Statistical Measures

```python
import numpy as np
import pandas as pd
from scipy.stats import skew, kurtosis
from scipy.signal import welch


st = pd.DataFrame()
# Statistical Measures
st['mean'] = df.iloc[:, 1:].mean(axis=1)
st['median'] = df.iloc[:, 1:].median(axis=1)
st['variance'] = df.iloc[:, 1:].var(axis=1)
st['std_dev'] = df.iloc[:, 1:].std(axis=1)
st['skewness'] = df.iloc[:, 1:].apply(skew, axis=1)
st['kurtosis'] = df.iloc[:, 1:].apply(kurtosis, axis=1)

st
```

frequency domain using welch

```python
# Frequency Domain Features using Welch's Method
def calculate_power_spectrum(signal):
    f, Pxx = welch(signal, fs=1, nperseg=len(signal))
    return f, Pxx


for i in range(len(df)):

    f, Pxx = calculate_power_spectrum(df.iloc[i, 1:])
    

    dominant_frequency = f[np.argmax(Pxx)]
    total_power = np.sum(Pxx)
    

    st['dominant_frequency'] =  dominant_frequency
    st['total_power'] = total_power

st
```

```python
import numpy as np

# Calculate the autocovariance for each time series in the DataFrame 'df'
autocovariance_list = []


for i in range(len(df)):
   
    time_series = df.iloc[i, 1:].values
    
    autocovariance = np.cov(time_series, ddof=0) 
    
  
    autocovariance_list.append(autocovariance)


autocovariance_array = np.array(autocovariance_list)

st['autocovariance'] = autocovariance_array.tolist()
```

```python
from sklearn.model_selection import cross_val_score

# Define a range of k values
k_values = [5, 10, 15, 20, 25, 30, 35, 40, 45, 50]
cv_scores = []

# Iterate over k values
for k in k_values:
    knn_classifier = KNeighborsClassifier(n_neighbors=k)
    scores = cross_val_score(knn_classifier, X_train, y_train, cv=5)  # 5-fold cross-validation
    cv_scores.append(scores.mean())

# Find the optimal k
optimal_k = k_values[cv_scores.index(max(cv_scores))]
print("Optimal k:", optimal_k)
```

```python
import numpy as np
import tensorflow as tf
from sklearn.model_selection import train_test_split
from tensorflow import keras
from tensorflow.keras import layers
from kerastuner.tuners import RandomSearch

# Assuming X contains your BCG data and y contains the corresponding labels
# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(df, y, test_size=0.2, random_state=42)

# Define the LSTM-based RNN model
def build_model(hp):
    model = keras.Sequential()
    model.add(layers.LSTM(
        units=hp.Int('units', min_value=32, max_value=128, step=32),
        input_shape=(96, 1)
    ))
    model.add(layers.Dense(1, activation='sigmoid'))
    
    model.compile(
        optimizer='adam',
        loss='binary_crossentropy',
        metrics=['accuracy']
    )
    return model

# Define the tuner
tuner = RandomSearch(
    build_model,
    objective='val_accuracy',
    max_trials=10,  # Number of hyperparameter combinations to try
    directory='tuner_directory',
    project_name='lstm_tuner'
)

# Perform the hyperparameter search
tuner.search(X_train, y_train, epochs=10, validation_data=(X_test, y_test))

# Get the best hyperparameters and build the model
best_hp = tuner.get_best_hyperparameters()[0]
best_model = tuner.hypermodel.build(best_hp)

# Train the best model
best_model.fit(X_train, y_train, epochs=10, validation_data=(X_test, y_test))

# Evaluate the best model
loss, accuracy = best_model.evaluate(X_test, y_test)
print("Loss:", loss)
print("Accuracy:", accuracy)
```

```python
X_train, X_test, y_train, y_test = train_test_split(df, y, test_size=0.2, random_state=42)

# Reshape the data to fit the CNN input shape
X_train = np.expand_dims(X_train, axis=-1)
X_test = np.expand_dims(X_test, axis=-1)

# Define the CNN model
def build_model(hp):
    model = keras.Sequential()
    model.add(layers.Conv1D(
        filters=hp.Int('filters', min_value=32, max_value=128, step=32),
        kernel_size=hp.Int('kernel_size', min_value=3, max_value=5, step=1),
        activation='relu',
        input_shape=(96, 1)
    ))
    model.add(layers.MaxPooling1D(
        pool_size=hp.Int('pool_size', min_value=2, max_value=3, step=1)
    ))
    model.add(layers.Flatten())
    model.add(layers.Dense(
        units=hp.Int('units', min_value=32, max_value=128, step=32),
        activation='relu'
    ))
    model.add(layers.Dense(1, activation='sigmoid'))
    
    model.compile(
        optimizer='adam',
        loss='binary_crossentropy',
        metrics=['accuracy']
    )
    return model

# Define the tuner
tuner = RandomSearch(
    build_model,
    objective='val_accuracy',
    max_trials=20,  # Number of hyperparameter combinations to try
    directory='tuner_directory',
    project_name='cnn_tuner'
)

# Perform the hyperparameter search
tuner.search(X_train, y_train, epochs=10, validation_data=(X_test, y_test))

# Get the best hyperparameters and build the model
best_hp = tuner.get_best_hyperparameters()[0]
best_model = tuner.hypermodel.build(best_hp)

# Train the best model
best_model.fit(X_train, y_train, epochs=10, validation_data=(X_test, y_test))

# Evaluate the best model
loss, accuracy = best_model.evaluate(X_test, y_test)
print("Loss:", loss)
print("Accuracy:", accuracy)

```
